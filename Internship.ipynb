Clearing the Datasets

For Raw Data

```python
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
raw = []
for i in range(0, 34116):
  cell = re.sub('[^a-zA-Z]', ' ', Raw_Data['RAW DATA'][i])
  cell = cell.lower()
  cell = cell.split()
  ps = PorterStemmer()
  all_stopwords = stopwords.words('english')
  all_stopwords.remove('not')
  cell = [ps.stem(word) for word in cell if not word in set(all_stopwords)]
  cell = ' '.join(cell)
  raw.append(cell)
```

For Train Data

```python
train_set = []
for i in range(0, 979):
  cell = re.sub('[^a-zA-Z]', ' ', Example_Data['Technology Skills'][i])
  cell = cell.lower()
  cell = cell.split()
  ps = PorterStemmer()
  all_stopwords = stopwords.words('english')
  all_stopwords.remove('not')
  cell = [ps.stem(word) for word in cell if not word in set(all_stopwords)]
  cell = ' '.join(cell)
  train_set.append(cell)
  
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 1500)
X = cv.fit_transform(train_set).toarray()
X_test = cv.transform(raw).toarray()
y = np.ones((979,))
```

Using Naive Bayes Model

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X, y)
```

Predicting on Raw Data


```python
select = classifier.predict(X_test)
```

```python
Selected=[]
```

```python
for i in range(len(X_test)):
    if select[i] == 1:
        Selected.append(X_test[i])
```

Converting the cv transformed data back to Readable Form

```python
Selected = cv.inverse_transform(Selected)
```

```python
Final_Words = []
```
